```python
import pandas as pd
```

Load dataframes

```python
df_test = pd.read_csv('data/datasets_full/social_spambots_1/users.csv', index_col=False, encoding='utf-8')
df_test
```

```python
df_ss1 = pd.read_csv('data/datasets_full/social_spambots_1/tweets.csv', index_col=False, encoding='utf-8')
df_ss2 = pd.read_csv('data/datasets_full/social_spambots_2/tweets.csv', index_col=False, encoding='utf-8')
df_ss3 = pd.read_csv('data/datasets_full/social_spambots_3/tweets.csv', index_col=False, encoding='utf-8')
df_ts1 = pd.read_csv('data/datasets_full/traditional_spambots_1/tweets.csv', index_col=False, encoding='utf-8')
df_ff = pd.read_csv('data/datasets_full/fake_followers/tweets.csv', index_col=False, encoding='utf-8')
df_ga = pd.read_csv('data/datasets_full/genuine_accounts/my_table.csv', delimiter='\t', encoding='latin1')
```

Remove columns and create one bot dataframe.

```python
df_ss1.drop(columns=['id',
                     'source',
                     'truncated',
                     'in_reply_to_status_id',
                     'retweeted_status_id',
                     'geo',
                     'contributors',
                     'favorited',
                     'retweeted',
                     'created_at',
                     'timestamp',
                     'crawled_at',
                     'updated'
                    ], inplace=True
           )

df_ss2.drop(columns=['id',
                     'source',
                     'truncated',
                     'in_reply_to_status_id',
                     'retweeted_status_id',
                     'geo',
                     'contributors',
                     'favorited',
                     'retweeted',
                     'created_at',
                     'timestamp',
                     'crawled_at',
                     'updated'
                    ], inplace=True
           )

df_ss3.drop(columns=['id',
                     'source',
                     'truncated',
                     'in_reply_to_status_id',
                     'retweeted_status_id',
                     'geo',
                     'contributors',
                     'favorited',
                     'retweeted',
                     'created_at',
                     'timestamp',
                     'crawled_at',
                     'updated'
                    ], inplace=True
           )

df_ts1.drop(columns=['id',
                     'source',
                     'truncated',
                     'in_reply_to_status_id',
                     'retweeted_status_id',
                     'geo',
                     'contributors',
                     'favorited',
                     'retweeted',
                     'created_at',
                     'timestamp',
                     'crawled_at',
                     'updated'
                    ], inplace=True
           )

df_ff.drop(columns=['id',
                     'source',
                     'truncated',
                     'in_reply_to_status_id',
                     'retweeted_status_id',
                     'geo',
                     'contributors',
                     'favorited',
                     'retweeted',
                     'created_at',
                     'timestamp'
                    ], inplace=True
           )

df = pd.concat([df_ss1, df_ss2, df_ss3, df_ts1, df_ff])

# create label for bots
df['bot'] = 1
```

Create actual users.

```python
df_ga.drop(columns=['id',
                     'source',
                     'truncated',
                     'in_reply_to_status_id',
                     'retweeted_status_id',
                     'geo',
                     'contributors',
                     'favorited',
                     'retweeted',
                     'created_at',
                     'timestamp',
                     'crawled_at',
                     'updated'
                    ], inplace=True
           )

df_ga['bot'] = 0
```

Add to main df and reset index.

```python
df = pd.concat([df, df_ga])

df.reset_index(drop=True, inplace=True)
df
```

Save dataframe to csv.

```python
df.to_csv('bot_data.csv')
```
